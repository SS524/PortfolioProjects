## Trying to find out the distribution of 'fnlwgt' column

```python
plt.figure(figsize=(13,13))
sns.histplot(df_copy['fnlwgt'],kde=True)
```

### From the above diagram, we understood that the distribution is following a Right Skewed distribution. Let's try to find out the outliers and remove those points from our dataframe 

```python
def identify_outliers(dataframe,columnName):
    sorted_list=list(dataframe[columnName].sort_values(ascending=False))
    res=[]
    q1=np.percentile(sorted_list,25)
    q3=np.percentile(sorted_list,75)
    IQR=q3-q1
    lower_bound=q1-(1.5*IQR)
    upper_bound=q3+(1.5*IQR)
    for i in sorted_list:
        if(i>upper_bound or i<lower_bound):
            res.append(i)
    return res
```

```python
len(identify_outliers(df_copy,'fnlwgt'))
```

## There are total 903 outliers for 'fnlwgt' column. Let's remove it

```python
for i in identify_outliers(df_copy,'fnlwgt'):
    df_copy=df_copy[~(df_copy['fnlwgt']==i)]
```

```python
df_copy
```

```python
plt.figure(figsize=(13,13))
sns.histplot(df_copy['fnlwgt'],kde=True)
```

### From the abouve diagram, we can see that the distribution has shifted towards the centre of the graph after removing the outliers

# Part 3: Data Visualization

## Finding the trends of income in the United States with respect to workclass

```python
df_workclass_salary=df_cleaned[df_cleaned['native-country']=='United-States'].groupby(['workclass','income']).count().reset_index()
```

```python
df_workclass_salary
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=df_workclass_salary,x='workclass',y='age',hue='income')
plt.ylabel('Count')
plt.title('Salary comparison for every work classes in the United Staes',fontsize=20)
plt.show()
```

### From the above diagram, we understood that the maximum number of employees, who earn more than 50k annually, are coming from private sector.

## Comparing Occupation with Sex

```python
genders=['Male','Female']

for gen in genders:
    d=dict(df_cleaned[df_cleaned['sex']==gen]['occupation'].value_counts())
    ex=np.zeros(len(d))
    ls=list(d.values())
    ind=ls.index(max(d.values()))
    ex[ind]=0.1
    plt.figure(figsize=(9,9))
    plt.pie(x=d.values(),labels=d.keys(),textprops={'fontsize':13},explode=ex,autopct='%1.1f%%')
    plt.title(f'{gen} Vs occupation',fontsize=20)
    plt.tight_layout()
    
    
    
```

### The most common occupations among males and females, are Craft-repair and Adm-cleirical respectively
